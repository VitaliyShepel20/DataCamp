# Defining a function counting a number of repeated value in the columns of DataFrame

```python
import pandas as pd

#creating pandas DataFrame
crimes_2020_df = pd.read_csv('C://Users//Vitaliy//Downloads//Vitaliy//Crimes_-_2020.csv')

#defining a function col_count
def count_entries(df, col_name):
    #initializing a new dict)
    col_count = {}
    
    #extract a col_name from crimes_2020_df: col
    col = crimes_2020_df[col_name]
    
    #loop over col 
    for entry in col:
        #if the entry in the dict.key, add 1
        if entry in col_count.keys():
            col_count[entry] += 1
            
        #else initialize the column count
        else:
            col_count[entry] = 1
            
    return col_count

#call the function count_entries
result = count_entries(crimes_2020_df, 'Location Description')
                  
print(result)
```
